```python
def display_general():
  print("General Information")
  print("---------------------------------")
  print("Please order for one person at a time, aged 4 or above")
  print("Suggested calorie guidelines:")
  print("\tMen (adult): 2200-3200")
  print("\tWomen (adult): 1600-2400")
  print("\tMen (teen): 2000-3200")
  print("\tWomen (teen): 1800-2400")
  print("\tBoys (9-13): 1600-2600")
  print("\tGirls (9-13): 1400-2200")
  print("\tBoys (4-8): 1200-2000")
  print("\tGirls (4-8): 1200-1800")
  print("---------------------------------")

  print("\nMenu")
  print("\t1. Burger (beef, lettuce, tomato and pickles on sesame seed bun) \u2014 400 calories")
  print("\t2. Chicken salad (chicken, mayo, onion, celery) \u2014 524 calories")
  print("\t3. Fish tacos (white fish, corn tortilla, cabbage) \u2014 130 calories")
  print("\t4. Hot dog (Polish wieners with potato bun) \u2014 310 calories")
  print("\t0. Exit")

def display_extras():
  print("\nExtras")
  print("\t1. Chips \u2014 180 calories")
  print("\t2. Fries \u2014 320 calories")
  print("\t3. Water or diet soda\u2014 0 calories")
  print("\t4. Soda \u2014 about 150 calories")
  print("\t5. Cookie \u2014 about 150 calories")
  print("\t0. None")
```

```python
import sys
print("THE LUNCHBREAK")
order = []
calories = 0
display_general()
selection = input("\nEnter selection(s) with spaces in between: ")
## no inputs or exit
if (selection.strip() == "" or "0" in selection.strip()):
  print("Exiting now")
  sys.exit()
selection = selection.split(" ")
if len(selection) > 0:
  for i in range(len(selection)):
    ## non-numeric input
    if (selection[i].isdigit() == False):
      print("Invalid input, exiting now")
      sys.exit()
    if (int(selection[i]) < 1 or int(selection[i]) > 4):
      print("Invalid input, exiting now\n\n")
      sys.exit()
    selection[i] = int(selection[i])
    if (int(selection[i]) == 1):
      print("Adding burger")
      order.append("Burger \u2014 400 calories")
      calories += 400
    elif (int(selection[i]) == 2):
      print("Adding chicken salad")
      order.append("Chicken salad \u2014 524 calories")
      calories += 524
    elif (int(selection[i]) == 3):
      print("Adding fish tacos")
      order.append("Fish tacos \u2014 130 calories")
      calories += 130
    elif (int(selection[i]) == 4):
      print("Adding hot dog")
      order.append("Hot dog \u2014 310 calories")
      calories += 310
display_extras()
side = input("Enter sides with spaces in between: ")
if (side.strip() == "" or "0" in side.strip()):
  print("No sides selected")
else:
  side = side.split(" ")
  for i in range(len(side)):
    ## check if numeric
    if (side[i].isdigit() == False):
      print("Invalid input, exiting now\n\n")
      sys.exit()
    side[i] = int(side[i])
    if (side[i] < 1 or side[i] > 5):
      print("Invalid input, exiting now\n\n")
      sys.exit()
    side[i] = int(side[i])
    if (side[i] == 1):
      print("Adding chips")
      order.append("Chips \u2014 180 calories")
      calories += 180
    elif (side[i] == 2):
      print("Adding fries")
      order.append("Fries \u2014 320 calories")
      calories += 320
    elif (side[i] == 3):
      print("Adding water or diet soda")
      order.append("Water/diet soda \u2014 0 calories")
    elif (side[i] == 4):
      print("Adding soda")
      order.append("Soda \u2014 150 calories")
      calories += 150
    elif (side[i] == 5):
      print("Adding cookie")
      order.append("Cookie \u2014 150 calories")
      calories += 150
print("")
## display itemized order
for i in range(len(order)):
  print(order[i])
print("Total calories: ", calories)

## get gender
gender = input("Enter gender (m/f): ")
## validate gender
while (gender.lower().strip() != "m" and gender.lower().strip() != "f"):
  gender = input("Invalid input. Enter gender (m/f); ")
## validate age
age = input("Enter age: ")
## validate age
while (age.isdigit() == False or int(age) < 0):
  age = input("Invalid input. Enter age: ")
age = int(age)
if gender == 'm':
  if age <= 8:
    if calories > 2000:
      print("This person's order exceeds their dietary guidelines. Consider ordering less next time around")
  elif age <= 13:
    if calories > 2600:
      print("This person's order exceeds their dietary guidelines. Consider ordering less next time around")
  elif age <= 19:
    if calories > 3200:
      print("This person's order exceeds their dietary guidelines. Consider ordering less next time around")
  else:
    if calories > 3200:
      print("This person's order exceeds their dietary guidelines. Consider ordering less next time around")
elif gender == 'f':
  if age <= 8:
    if calories > 1800:
      print("This person's order exceeds their dietary guidelines. Consider ordering less next time around")
  elif age <= 13:
    if calories > 2200:
      print("This person's order exceeds their dietary guidelines. Consider ordering less next time around")
  elif age <= 19:
    if calories > 2400:
      print("This person's order exceeds their dietary guidelines. Consider ordering less next time around")
  else:
    if calories > 2400:
      print("This person's order exceeds their dietary guidelines. Consider ordering less next time around")
```
